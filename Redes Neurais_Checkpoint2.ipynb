```python
import pandas as pd 
#importa para a leitura do dataframe
```

```python
#lendo o dataframe diretamente do site fornecido
imoveis = pd.read_csv('https://tinyurl.com/alugueis-sp-df') 
imoveis.head(5) 
```

```python
imoveis.info
```

```python
imoveis[imoveis["Negotiation Type"]=="sale"] #observando apenas os apartamentos que estão a venda(sale)
```

```python
imoveis[imoveis["Property Type"]=="apartment"]  #Observando se tem algo além de apartment
#no caso, foi concluido que só possui apartment no dataframe
```

```python
#dropando colunas desnecessárias
df = imoveis.drop(imoveis[imoveis["Negotiation Type"] != "sale"].index) #aqui foi deixado apenas os dados que contém sale
df = df.drop("Latitude", axis=1)
df = df.drop("Longitude", axis=1)
```

```python
#dropando colunas que tinham a mesma informação em todas as linhas
df = df.drop("Negotiation Type", axis=1)
df = df.drop("Property Type", axis=1)
```

```python
df.shape
```

```python
df
```

```python
df2 = df
```

```python
df2 = df2.drop("District", axis=1)
```

```python
df[["Bairros", "Cidade"]] = df['District'].str.split('/', expand=True) #separação de District, fazendo duas colunas novas
```

```python
#Só tem bairros da cidade de São Paulo, então resolvi dropar a coluna Cidade e District, deixando apenas os bairros
df = df.drop("District", axis=1)
df = df.drop("Cidade", axis=1)
```

```python
df #observando como ficou o dataframe
```

```python
#convertendo dados categóticos em 0 ou 1
xy = pd.get_dummies(df['Bairros'])
df = pd.concat([df,xy], axis=1)
```

```python
df = df.drop("Bairros", axis=1)
```

```python
df
```

## Criação dos modelos

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
#separando dados e label
X = df.drop(columns = ["Price"])
Y = df["Price"]
```

```python
#20%
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=38)
```

```python
scaler = StandardScaler() #escalando os dados, fazendo a normalização
x_train_scaled = scaler.fit_transform(X_train)
x_test_scaled = scaler.transform(X_test)
```

```python
#Modelo 1
p1 = Perceptron(random_state=42)
p1.fit(x_train_scaled, Y_train)
Y_pred = p1.predict(x_test_scaled)
```

```python
#Modelo 2
ml1 = MLPRegressor(hidden_layer_sizes=(15,), random_state = 38, max_iter = 100, solver='adam')
ml1.fit(x_train_scaled, Y_train)

print(f"Mean Squared Error : {round(mean_squared_error(ml1.predict(x_test_scaled), Y_test), 2)}")
print(f"Mean Absolute Error : {round(mean_absolute_error(ml1.predict(x_test_scaled), Y_test), 2)}")
print(f"R2 Score : {r2_score(ml1.predict(x_test_scaled), Y_test)}")
```
